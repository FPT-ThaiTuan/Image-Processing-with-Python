```python
import cv2
from google.colab.patches import cv2_imshow
```

* The code imports the cv2 library for image processing in Python.
  1. It loads images from a file using cv2.imread().
  2. It then displays the image in the window with cv2.imshow()
  3. cv2.waitKey(0) pauses execution until a key is pressed.
  4. Finally, cv2.destroyAllWindows() closes all windows.






```python
# Loading image from a file
image = cv2.imread('./Data/test.jpg')

# Displaying an image in a window
print('--------------------Image origin-----------------\n')
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 2. Converting an image to grayscale form

* The code imports the cv2 library for image processing in Python.
  1. cv2.cvtColor() converts the loaded image into a grayscale.




```python
# Converting an image to grayscale form
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Displaying the grayscale image
print('-------------Gray Image\n------------------')
cv2_imshow(gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 3. Edge Detection

* The code imports the cv2 library for image processing in Python.
  1. You can use the Canny edge detection algorithm in OpenCV with the cv2.Canny() function to detect edges in a grayscale image.



```python
# Applying Canny edge detection (a simple edge detection method)
edges = cv2.Canny(gray_image, 50, 150)

# Displaying the edge-detected image
print('-----------------Edges----------------\n')
cv2_imshow(edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. Resize and rotate images

* The code imports the cv2 library for image processing in Python.
  1. cv2.resize(): Resizes an image to a specified width and height.
  2. cv2.getRotationMatrix2D(): Computes a rotation matrix for a given rotation angle, center point, and scale factor.
  3. cv2.warpAffine(): Applies an affine transformation to an image using the provided transformation matrix.

```python
width, height = 300, 200

# Resizing the image
resized_image = cv2.resize(image, (width, height))

# Rotating the image by 45 degrees
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), 45, 1)
rotated_image = cv2.warpAffine(resized_image, rotation_matrix, (width, height))

# Displaying the resized and rotated images
print('----------------Resized Image ( 300 * 200 ) -------------\n')
cv2_imshow(resized_image)
print('\n----------------Rotated Image-------------\n')
cv2_imshow(rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 5. Image blurring

* The code imports the cv2 library for image processing in Python.
  1. cv2.GaussianBlur(): This function applies a Gaussian filter to blur the image. It helps reduce noise in images and smooth edges.
  2. cv2.medianBlur(): This function applies a median filter to blur the image. It also helps reduce noise but preserves edge features.
  3. cv2.bilateralFilter(): This function applies a bilateral filter to blur the image. It retains both edge characteristics and noise reduction.

```python
import matplotlib.pyplot as plt
```

```python
# Apply different blurring techniques
gaussian = cv2.GaussianBlur(image, (7, 7), 0)
median = cv2.medianBlur(image, 5)
bilateral = cv2.bilateralFilter(image, 9, 75, 75)

# Convert images from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gaussian_rgb = cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB)
median_rgb = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)
bilateral_rgb = cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB)

# Create a figure with 1 row and 5 columns
plt.figure(figsize=(15, 5))

# Plot original image
plt.subplot(1, 5, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Plot Gaussian blur image
plt.subplot(1, 5, 2)
plt.imshow(gaussian_rgb)
plt.title('Gaussian Blur')
plt.axis('off')

# Plot median blur image
plt.subplot(1, 5, 3)
plt.imshow(median_rgb)
plt.title('Median Blur')
plt.axis('off')

# Plot bilateral blur image
plt.subplot(1, 5, 4)
plt.imshow(bilateral_rgb)
plt.title('Bilateral Blur')
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```

## 7.  Flipping an image

* The code imports the cv2 library for image processing in Python.
  1. cv2.flip(image, 1): This line flips the image horizontally. The second argument 1 specifies that the image should be flipped around the y-axis (horizontal flip).

  2. cv2.flip(image, 0): This line flips the image vertically. The second argument 0 specifies that the image should be flipped around the x-axis (vertical flip).

  3. cv2.flip(image, -1): This line flips the image both horizontally and vertically. The second argument -1 specifies that the image should be flipped around both x-axis and y-axis.

```python
# Flip the image horizontally
flipped_horizontal = cv2.flip(image, 1)

# Flip the image vertically
flipped_vertical = cv2.flip(image, 0)

# Flip the image both horizontally and vertically
flipped_both = cv2.flip(image, -1)

# Convert images from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
flipped_horizontal_rgb = cv2.cvtColor(flipped_horizontal, cv2.COLOR_BGR2RGB)
flipped_vertical_rgb = cv2.cvtColor(flipped_vertical, cv2.COLOR_BGR2RGB)
flipped_both_rgb = cv2.cvtColor(flipped_both, cv2.COLOR_BGR2RGB)

# Create subplots for original and flipped images
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(flipped_horizontal_rgb)
plt.title('Flipped Horizontal')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(flipped_vertical_rgb)
plt.title('Flipped Vertical')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(flipped_both_rgb)
plt.title('Flipped Both')
plt.axis('off')

plt.show()
```

##  8. Inverting an image

* The code imports the cv2 library for image processing in Python.
  1. cv2.bitwise_not(image) performs the bitwise NOT operation on each pixel of the image, effectively inverting the colors.

```python
image_dog = cv2.imread('./Data/test_invert.jpg')

invert_img = cv2.bitwise_not(image_dog)

# Display the original and inverted images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_dog)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(invert_img)
plt.title('Inverted Image')
plt.axis('off')

plt.show()
```

## 9. Image Masking

* The code imports the cv2 library for image processing in Python.
  1. cv2.circle(): This function is used to draw a circle or a filled circle on an image or a mask. Input parameters include:
    1. img: Image or mask to draw a circle on.
    2. center: Center coordinates of the circle.
    3. radius: Radius of the circle.
    4. color: The color of the circle.
    5. thickness: Thickness of the drawn line. If it is a negative value or cv2.FILLED, the circle will be filled.
  2. cv2.bitwise_and(): This function performs bitwise AND between two images. It applies a mask to an image to retain only pixels with non-zero values in both images. Input parameters include:
    1. src1: First input image.
    2. src2: Second input image.
    3. mask: Mask is applied to the src1 image.

```python
import numpy as np
```

```python
# Create a black mask with the same size as the image
mask = np.zeros_like(image_dog)
# Define the center and radius of the circle
height, width = image_dog.shape[:2]
center = (width // 2, height // 2)
radius = min(center[0], center[1])

# Draw a white circle in the mask
cv2.circle(mask, center, radius, (255, 255, 255), -1)

# Apply the mask to the image
masked_image = cv2.bitwise_and(image_dog, mask)

plt.subplot(1, 2, 1)
plt.imshow(image_dog)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(masked_image)
plt.title('Masked Image')
plt.axis('off')

plt.show()
```

## 10. Grayscale Histogram

* Grayscale histograms are valuable tools for analyzing the distribution of pixel intensities in grayscale images. They provide a visual summary of the frequency of occurrence of each pixel intensity value, aiding in tasks like thresholding, image enhancement, and segmentation in image processing.

```python
# Plotting the histogram of a grayscale image
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.hist(gray_img.ravel(), 256, [0, 256])
plt.title("Grayscale Histogram")
plt.show()

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calculate the grayscale histogram using cv2.calcHist()
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

# Display the histogram using matplotlib
plt.plot(hist, color='gray')
plt.xlabel('Intensity Level')
plt.ylabel('Pixel Count')
plt.title('Grayscale Histogram')
plt.show()
```

## 11. Color Histogram

* Color histogram: Graphical representation of color distribution in an image, aiding analysis of color content. OpenCV offers functions to compute color histograms, useful for segmentation, recognition, and retrieval. Colors are represented in RGB or HSV color space, with RGB as combinations of red, green, and blue, and HSV as hue, saturation, and value components.

```python
color = ('r', 'g', 'b')
for i, col in enumerate(color):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])

plt.title("Color Histogram")
plt.show()
```
